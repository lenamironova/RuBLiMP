```python
import glob, pandas as pd

paths = glob.glob("./generated_data/argument_structure/*.tsv")
print("Number of files:", len(paths))
assert paths, "No files in ./generated_data/argument_structure/"

df = pd.concat((pd.read_csv(p, sep="\t") for p in paths), ignore_index=True)
print("Number of pairs:", len(df))
```

```python
import pandas as pd

pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_rows", 50)

df_ins_obj = df[df["phenomenon_subtype"].str.contains("ins_obj")].copy()
print("Number of pairs:", len(df_ins_obj))
df_ins_obj[["source_sentence", "target_sentence"]].sample(10, random_state=42)

```

```python
out_ins_obj = "../generated_data/argument_structure_ins_obj_only.csv"
df_ins_obj.to_csv(out_ins_obj, index=False, encoding="utf-8")
out_ins_obj
```
